```python
def helper(inp):
    DT = inp.split(':')
    TOP = float(DT[0]) + (float(DT[1])/60)
    return TOP
```

```python
df.TimeOfPossession = df.TimeOfPossession.map(lambda x: helper(x))
```

```python
X = df.copy()
y = df['Score'].values.reshape(-1, 1)
```

```python
X.shape, y.shape
```

```python
X = pd.get_dummies(X,columns=['PlayingSurface', 'Opponent', 'Team', 'HomeOrAway'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
