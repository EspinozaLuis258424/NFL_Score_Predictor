Decided to run multiple predictions and tests to plot out the score differential and mean accuracy value

```python
# Run algorithm 100 times, capturing the test/prediction values as data for the graphs
# algorithm function
def randomForestRunner(num_iterations):

    results_list = []

    for n in range(1, num_iterations+1):
        X_train, X_test, y_train, y_test = train_test_split(X, y)
        rf = RandomForestClassifier(n_estimators=1000)
        rf = rf.fit(X_train, y_train.ravel())
        prediction = rf.predict(X_test[0:1].values)
        test = y_test[0]

        results = {
            'iteration#': n,
            'score': rf.score(X_test, y_test),
            'prediction': prediction,
            'test': test
        }

        results_list.append(results)
        
    return results_list
```

```python
results_list = randomForestRunner(100)
```

### Visualizations

```python
# imports
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Configure X and Y Values, labels
x_data = []
y_data = []
labels = []

for obj in results_list:
    x_data.append(obj['prediction'][0] - obj['test'][0])
    y_data.append(obj['score'])
    labels.append(obj['iteration#'])

y_data
```

```python
# scatter plot -> bubbles/dots will represent each iteration and their respective scores vs
plt.scatter(x=x_data, y=y_data, marker='o')
plt.xlabel('Score Diff (Prediction & Test)') 
plt.ylabel('Mean Accuracy Value') 
plt.show()
```
